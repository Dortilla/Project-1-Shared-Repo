# Analysis and Findings

```python
# Import dependencies
import pandas as pd
from matplotlib import pyplot as plt
from scipy import stats as st
```

## Suicide Rate and Country Unemployment Rate

```python
# Load df from exploration.ipynb
sui_vs_uem = pd.read_csv('suicide_vs_unemployment_clean')

# Create two series, one for suicide rates and one for uemployment rate
sui = sui_vs_uem['suicide rate']
uem = sui_vs_uem['unemployment rate']
```

```python
# Linnear Regression
m, b, r, p, e = st.linregress(uem, sui)
sui_pre = m * uem + b
r_2_str = "{}\u00b2".format('r')
sui_pre_str = f'y={round(m,2)}*x+{round(b,2)}, pvalue={"{:.2e}".format(p)}, {r_2_str}={round(r*r,2)})'

# Scatter plot
plt.scatter(uem, sui, s=2)
plt.title('Suicide vs Unemployment for Countries')
plt.xlim(0, max(uem))
plt.xlabel('Unemployment Rate')
plt.ylim(0,60)
plt.ylabel('Suicide Rate (Age Normalized, per 1000 People)')
plt.plot(uem, sui_pre, c='red')
plt.annotate(sui_pre_str, (5,50))
plt.show()
```

## Suicide Rate and Country GDP

```python
# Load df from exploration.ipynb
sui_vs_gdp = pd.read_csv('suicide_vs_gdp_clean')

# Create two series, one for suicide rates and one for uemployment rate
sui = sui_vs_gdp['suicide rate']
gdp = sui_vs_gdp['gdp per capita']
```

```python
# Linnear Regression
m, b, r, p, e = st.linregress(gdp, sui)
gdp_pre = m * gdp + b
r_2_str = "{}\u00b2".format('r')
gdp_pre_str = f'y={round(m,2)}*x+{round(b,2)}, pvalue={"{:.2e}".format(p)}, {r_2_str}={round(r*r,2)})'

# Scatter plot
plt.scatter(gdp, sui, s=2)
plt.title('Suicide vs GDP per capita for Countries')
plt.xlim(0, max(gdp))
plt.xlabel('GDP Rate')
plt.ylim(0,60)
plt.ylabel('Suicide Rate (Age Normalized, per 1000 People)')
plt.plot(gdp, gdp_pre, c='red')
plt.annotate(gdp_pre_str, (35000, 30))
plt.show()
```
